#### 5) Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
estGranada = dataset[dataset['Provincia']=='Granada'].groupby('Municipio').agg(suma_poblacion = ('Poblacion','sum'))
estGranada['suma_poblacion']
print('Los datos de la población de Granada son:')
print('Media:', estGranada['suma_poblacion'].mean())
print('Mediana:', estGranada['suma_poblacion'].median())
print('Desviación estándar:', estGranada['suma_poblacion'].std())
print('Mínimo', estGranada['suma_poblacion'].min())
print('Máximo:', estGranada['suma_poblacion'].max())
```

#### 6) Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
gpddataset = dataset.groupby(['Provincia','Municipio']).agg(suma_poblacion = ('Poblacion','sum'))
for provincia, datos in gpddataset.groupby(level=0):
    sns.histplot(datos['suma_poblacion'],kde=True)
    plt.title('Histograma de Población en Municipios de {}'.format(provincia))
    plt.xlabel('Población')
    plt.ylabel('Frecuencia')
    plt.show()
```

#### 7) Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.

```python
gmdataset =dataset.groupby('Municipio').agg(suma_poblacion = ('Poblacion','sum'))
rndMuni= gmdataset.sample(n=50, replace=False)
print('De los 50 municipios seleccionados alatoriamente su media y desviación estándar es la siguiente:')
print('Media:', rndMuni['suma_poblacion'].mean())
print('Desviación estándar:', rndMuni['suma_poblacion'].std())
print('Los municipios se detallan en la lista siguiente:',rndMuni.reset_index()['Municipio'].tolist())
```

#### 8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
gmdataset =dataset.groupby('Municipio').agg(suma_poblacion = ('Poblacion','sum'))
madrid_data = dataset[dataset['Municipio'] == 'Madrid']
coord_madrid = (madrid_data['X'].iloc[0],madrid_data['Y'].iloc[0])
print('Las coordenadas de Madrid son',coord_madrid)
```

```python

datasetnoMadrid = dataset[dataset['Municipio']!='Madrid'].copy()

datasetnoMadrid.loc[:,'Distancia_a_Madrid'] = np.sqrt((datasetnoMadrid['X'] - coord_madrid[0])**2 + (datasetnoMadrid['Y'] - coord_madrid[1])**2)

orden_cercano = datasetnoMadrid.sort_values(by='Distancia_a_Madrid')
orden_lejano = datasetnoMadrid.sort_values(by='Distancia_a_Madrid', ascending=False)

orden_cercano_unicos = orden_cercano.drop_duplicates(subset=['Municipio'])
orden_lejano_unicos = orden_lejano.drop_duplicates(subset=['Municipio'])
# Obtener los 5 municipios más cercanos a Madrid
municipios_mas_cercanos = orden_cercano_unicos.head(5)

# Obtener los 5 municipios más lejanos a Madrid
municipios_mas_lejanos = orden_lejano_unicos.tail(5)

print("Los 5 municipios más cercanos a Madrid:")
print(municipios_mas_cercanos[['Municipio', 'Distancia_a_Madrid']])
print("\nLos 5 municipios más lejanos a Madrid:")
print(municipios_mas_lejanos[['Municipio', 'Distancia_a_Madrid']])
```
